## Trực quan hóa kết quả độ trễ (thời gian phản hồi)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ["LogisticRegression", "Random Forest", "SVM"]
mean_full = [1.917, 4.173, 2.858]
p50_full = [1.872, 4.028, 2.443]
p95_full = [2.330, 4.965, 5.136]
mean_pca = [0.650, 2.982, 0.925]
p50_pca = [0.627, 2.914, 0.821]
p95_pca = [0.772, 3.451, 1.641]
mean_reduced = [0.936, 3.232, 1.280]
p50_reduced = [0.903, 3.165, 1.253]
p95_reduced = [1.118, 3.676, 1.765]

x = np.arange(len(models))
width = 0.1

plt.figure(figsize=(10, 6))

# Full features bars
plt.bar(x - 3*width, mean_full, width, label="Full Mean")
plt.bar(x - 2*width, p50_full, width, label="Full P50")
plt.bar(x - width, p95_full, width, label="Full P95")

# PCA features bars
plt.bar(x, mean_pca, width, label="PCA Mean")
plt.bar(x + width, p50_pca, width, label="PCA P50")
plt.bar(x + 2*width, p95_pca, width, label="PCA P95")

# Reduced features bars
plt.bar(x + 3*width, mean_reduced, width, label="Reduced Mean")
plt.bar(x + 4*width, p50_reduced, width, label="Reduced P50")
plt.bar(x + 5*width, p95_reduced, width, label="Reduced P95")

plt.xticks(x + width, models, rotation=30, ha="right")
plt.ylabel("Latency (ms)")
plt.title("Latency Metrics (Mean, P50, P95) Across Models and Feature Sets")
plt.legend(ncol=3, fontsize='small')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ["LogisticRegression", "Random Forest", "SVM"]
mean_full = [1.917, 4.173, 2.858]
p50_full = [1.872, 4.028, 2.443]
p95_full = [2.330, 4.965, 5.136]
mean_pca = [0.650, 2.982, 0.925]
p50_pca = [0.627, 2.914, 0.821]
p95_pca = [0.772, 3.451, 1.641]
mean_reduced = [0.936, 3.232, 1.280]
p50_reduced = [0.903, 3.165, 1.253]
p95_reduced = [1.118, 3.676, 1.765]

x = np.arange(len(models))

plt.figure(figsize=(10, 6))

# Plotting lines for Full features
plt.plot(x, mean_full, marker='o', label="Full Mean")
plt.plot(x, p50_full, marker='s', label="Full P50")
plt.plot(x, p95_full, marker='^', label="Full P95")

# Plotting lines for PCA features
plt.plot(x, mean_pca, marker='o', linestyle='--', label="PCA Mean")
plt.plot(x, p50_pca, marker='s', linestyle='--', label="PCA P50")
plt.plot(x, p95_pca, marker='^', linestyle='--', label="PCA P95")

# Plotting lines for Reduced features
plt.plot(x, mean_reduced, marker='o', linestyle=':', label="Reduced Mean")
plt.plot(x, p50_reduced, marker='s', linestyle=':', label="Reduced P50")
plt.plot(x, p95_reduced, marker='^', linestyle=':', label="Reduced P95")

plt.xticks(x, models, rotation=30, ha="right")
plt.ylabel("Latency (ms)")
plt.title("Latency Metrics (Mean, P50, P95) by Model and Feature Set")
plt.legend(fontsize='small', ncol=3)
plt.tight_layout()
plt.show()
```
