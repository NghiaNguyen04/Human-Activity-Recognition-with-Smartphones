# Label Encoding cho cột target (Activity)

## Load dữ liệu

```python
import pandas as pd

data_train = pd.read_csv('../../data/raw/train.csv')
data_test = pd.read_csv('../../data/raw/test.csv')
data_train.head()
```

## Sử dụng Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
data_train['Activity_code'] = labelEncoder.fit_transform(data_train.iloc[:, -1])
data_test['Activity_code'] = labelEncoder.transform(data_test.iloc[:, -1])
data_test.head()
```

## Lưu file

```python
# Lưu data_train dưới dạng CSV
data_train.to_csv('../../data/interim/train_LabelEncoder.csv', index=False)

# Lưu data_test dưới dạng CSV
data_test.to_csv('../../data/interim/test_LabelEncoder.csv', index=False)

print("Dữ liệu đã được lưu vào thư mục ../../data/interim")
```
