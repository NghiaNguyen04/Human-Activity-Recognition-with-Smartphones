```python
import numpy as np
from typing import List

def preprocess_input_LSTM(input_str: str) -> np.ndarray:
    elements = input_str.split()  # Tách chuỗi theo khoảng trắng/tab
    features = [float(e) for e in elements]  # Chuyển sang số thực
    return np.array(features)  # Trả về mảng numpy 1 chiều

# Chuẩn bị dữ liệu đầu vào
input_data = preprocess_input_LSTM(text)

# Thêm chiều để phù hợp với yêu cầu của LSTM (batch_size, timesteps, features)
input_data = np.expand_dims(input_data, axis=0)  # Thêm chiều cho batch_size (1 mẫu)
input_data = np.expand_dims(input_data, axis=1)  # Thêm chiều cho features (số đặc trưng)

# Dự đoán từ mô hình
output = model.predict(input_data)
```

```python
# Lấy index của giá trị xác suất lớn nhất
index_of_max = np.argmax(output, axis=-1)  # Tìm giá trị lớn nhất ở trục cuối cùng (trục features)

# Vì output có dạng (1, 1, 6), index_of_max sẽ có dạng (1, 1), nên cần lấy phần tử đầu tiên
index_of_max = index_of_max[0, 0]

print("Index of the highest probability:", index_of_max)
```
